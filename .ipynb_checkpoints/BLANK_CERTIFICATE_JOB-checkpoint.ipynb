```python
from PIL import Image, ImageDraw, ImageFont
import pandas as pd
import numpy as np
from matplotlib.pyplot import imshow
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
im=Image.open("D:\\certificate_blank.jpg")
```

```python
data=pd.read_excel('D:\\banquet.xlsx',)
data.index = np.arange(1, len(data) + 1)

data
```

```python
for x in range(1,(data.shape[0]-30)):
    im=Image.open("D:\\certificate_blank.jpg")
    print(x)
    draw = ImageDraw.Draw(im)
    fontsize = 100
    fnt = ImageFont.truetype("D:\\alex-brush\\AlexBrush-Regular.ttf",100)
    draw.text((400-20,1114-100), data.loc[x,'First Name'], font=fnt, fill=(212,175,55,0))
    # draw text, full opacity
    draw.text((1340-80,1067-50), data.loc[x,'Last Name'], font=fnt, fill=(212,175,55,0))
    im.save(f'D:\\ccertificatee{x}', "PNG")
    Image.open(f"D:\\ccertificatee{x}")
#     img = cv2.imread('D:\\ccertificatee.png')
#     cv2.imshow('students',img)
#     cv2.waitKey(0)

```
