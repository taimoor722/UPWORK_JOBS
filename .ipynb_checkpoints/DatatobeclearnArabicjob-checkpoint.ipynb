```python
import pandas as pd
# pd.set_option('display.max_rows', 660)
```

```python
raw=pd.read_excel("./Data to be cleaned from Survey Monkey.xlsx",header=0);
raw.iloc[:,9:11]
raw.loc[:,['الحالة الأجتماعية','Unnamed: 32','Unnamed: 33','Unnamed: 34','Unnamed: 35']]
```

```python
pd.set_option('display.max_rows', 660)
Y_N=raw.iloc[1:,9:11].replace('لأ','N').replace('نعم','Y')
```

```python
import numpy as np
Y_N['Would_Participate']=Y_N.iloc[:,1].replace(np.nan,'Y') 
```

```python
Y_N['Would_Participate']
```

```python
#M_F male female
# M_F['Male','Female']=
# M_F['Female']=['Unnamed: 22']
# raw['الجنس']
# M_F
M_F=raw.loc[1:,['الجنس','Unnamed: 22']]
M_F['Gender']=raw.loc[1:,'الجنس'].replace('ذكر','M').replace(np.nan,'F')
M_F.rename({'الجنس':'Male','Unnamed: 22':'Female'},axis='columns',inplace=True)
M_F
```

```python
#"Kuwaiti" or "Non-Kuwaiti"
K_NK=raw.loc[1:,['الجنسية','Unnamed: 24']]
K_NK['Nationality']=raw.loc[1:,'الجنسية'].replace('كويتي','Kuwaiti').replace(np.nan,'Non-Kuwaiti').replace('انثي','Non')
K_NK.rename({'الجنسية':'Kuwaiti','Unnamed: 24':'Non-Kuwaiti'},axis='columns',inplace=True)
K_NK
```

```python
#MatertialStat  
             #"Single",  "Married",  "Divorsed",  " Widow" or     "Separated"
matertial_stat=raw.loc[1:,['الحالة الأجتماعية','Unnamed: 32','Unnamed: 33','Unnamed: 34','Unnamed: 35']].replace('اعزب','Single').replace('متزوج','Married').replace('مطلق','Divorsed').replace('ارمل','Widow').replace('منفصل','Separated')
matertial_stat.rename({'الحالة الأجتماعية':"Single",'Unnamed: 32':"Married",'Unnamed: 33':"Divorsed",'Unnamed: 34':"Widow" ,'Unnamed: 35':"Separated"},inplace=True,axis=1)



matertial_stat['Matertial_stat']=np.nan
for x in ["Single",  "Married",  "Divorsed",  "Widow" ,"Separated"]:

    check=matertial_stat[x].notnull() & matertial_stat['Matertial_stat'].isnull()


    matertial_stat.loc[check,'Matertial_stat']=matertial_stat.loc[check,x]

    
matertial_stat
```

```python
#DO_YOU_DO_BUSINESS Y_N
raw.iloc[:,36:38]
Do_Business=raw.iloc[1:,36:38]
Do_Business['Did_You_Do_Business']=raw.iloc[1:,36].replace('نعم','Y').replace(np.nan,'N').replace('لأ','N')
Do_Business.rename({"هل قمت بتجارة ما في السابق ( مشاريع صغيرة منزلية أو فتحت محل على سبيل المثال) ؟‏":'Yes','Unnamed: 37':'No'},axis='columns',inplace=True)
Do_Business
```

```python
#Still_In-Business
raw.iloc[:,39:41]
Still_In_Business=raw.iloc[1:,39:41]
Still_In_Business['Still_In_Business']=raw.iloc[1:,39].replace('نعم','Y').replace(np.nan,'N').replace('لأ','N')
Still_In_Business.rename({"هل لازلت تملك مشروعا ؟":'Yes','Unnamed: 40':'No'},axis='columns',inplace=True)
Still_In_Business
```

```python
#How_Is_Buss
#"Weak","Good","VeryGood", or "Excellent"

raw.iloc[:,42:46]
How_Is_Buss=raw.loc[1:,['كيف تقيم نجاح مشروعك','Unnamed: 43','Unnamed: 44','Unnamed: 45']].replace('ضعيف','Weak').replace('جيد','Good').replace('جيد جدا','VeryGood').replace('ممتاز','Excellent')
How_Is_Buss.rename({'كيف تقيم نجاح مشروعك':"Weak",'Unnamed: 43':"Good",'Unnamed: 44':"VeryGood",'Unnamed: 45':"Excellent"},inplace=True,axis=1)



How_Is_Buss['How_Is_Buss']=np.nan
for x in ["Weak","Good","VeryGood","Excellent"]:

    check=How_Is_Buss[x].notnull() & How_Is_Buss['How_Is_Buss'].isnull()
    print(check)
    print(How_Is_Buss.loc[check,x])
    How_Is_Buss.loc[check,'How_Is_Buss']=How_Is_Buss.loc[check,x]
    print(How_Is_Buss)
    
    
How_Is_Buss
```

```python
#Would_You_In_Future
raw.iloc[:,46:48]
Would_You_In_Future=raw.iloc[1:,46:48]
Would_You_In_Future['Would_You_In_Future']=raw.iloc[1:,46].replace('نعم','Y').replace(np.nan,'N').replace('لأ','N')
Would_You_In_Future.rename({"هل تود ان تملك مشروعا او تتوسع في المستقبل؟":'Yes','Unnamed: 47':'No'},axis='columns',inplace=True)
Would_You_In_Future
```

```python
Q1=raw.iloc[1:,48:53]
Q1.replace(np.nan,0,inplace=True)
Q1
```

```python
# Q1['sum']=Q1.iloc[:,0]+Q1.iloc[:,1]+Q1.iloc[:,2]+Q1.iloc[:,3]+Q1.iloc[:,4]

def fun(x):
    xx=0
    c=0
    for x in x.tolist():
        if x!=0:
            c+=1
            xx+=x
        else:
            pass
    if c==0 or c==1:
        return xx
    else:
        s='average'
        return xx/c,s
Q1['average']=Q1.apply(fun,axis=1)
Q1
```

```python
Q2=raw.iloc[1:,53:58]
Q2.replace(np.nan,0,inplace=True)
Q2
```

```python
Q2['average']=Q2.apply(fun,axis=1)
Q2
```
