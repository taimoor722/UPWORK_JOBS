```python
Q1=raw.iloc[1:,48:53]
Q1.replace(np.nan,0,inplace=True)
Q1
```

```python
# Q1['sum']=Q1.iloc[:,0]+Q1.iloc[:,1]+Q1.iloc[:,2]+Q1.iloc[:,3]+Q1.iloc[:,4]

def fun(x):
    xx=0
    c=0
    for x in x.tolist():
        if x!=0:
            c+=1
            xx+=x
        else:
            pass
    if c==0 or c==1:
        return xx
    else:
        s='average'
        return xx/c,s
Q1['average']=Q1.apply(fun,axis=1)
Q1
```
