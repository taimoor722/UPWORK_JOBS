### Replacing emojis in text colum  with given corrosponding character

```python
emoji=pd.read_excel('D:\\emoji table (1).xlsx')
emoji[['Unnamed: 0','Unnamed: 2']]
```

```python
def fun0(x):                                              #using fun0 function to replace emoji with there similar=>[word expression] given in=> file emoji table (1).xlsx
    for em in emoji['Unnamed: 0'].values.tolist():
        if em in x:
            cha=emoji.loc[emoji['Unnamed: 0']==em,'Unnamed: 2']
            x=x.replace(em,str(str(cha.tolist()[0])))
            
    return x
data['new']=data['Text'].apply(fun0)
```

# spliting each emoji in each row and put them in list items for each row

```python
def emoji_seperation(each):
    print('each',each,'each')
    if re.search(r'[🇦-🇿]+',each,re.IGNORECASE):
        # print('if block started')
        # print('searching only a-z ',re.search(r'[🇦-🇿]+',each,re.IGNORECASE) )
        
        only_small_alpha=re.sub(r'[^🇦-🇿]+','',each)
        no_alpha=re.sub(r'[🇦-🇿]+','',each)
        no_alpha=list(no_alpha)
        # print('only small alpha :',only_small_alpha)
        # print('no alpha :',no_alpha)
        start=0
        for go in range(start,len(only_small_alpha),2):
            # print('grouping 2',only_small_alpha[go:go+2])
            no_alpha.append(only_small_alpha[go:go+2])
            # print('grouping listing :',no_alpha,end='  ')
            start=go+1
        print('\n============\n')
        # print(no_alpha)
        # all_emoji_present_unique.extend(no_alpha)
        # print(all_emoji_present_unique,len(all_emoji_present_unique),'length')
        return no_alpha
    else:
        # print(each,' :else block')
        all_emoji_present_unique.extend(each)
        # print('all_emoji_present_unique',len(all_emoji_present_unique),'length')
        return list(each)
all_emoji_present_unique=[]
print('====final===')
print(len(all_emoji_present_unique))
print(all_emoji_present_unique[:100])
print(re.sub(r'[^🇦-🇿]+','',' '.join(set(all_emoji_present_unique ))))
data['emoji_clean']=data['emoji'].apply(emoji_seperation)
```
