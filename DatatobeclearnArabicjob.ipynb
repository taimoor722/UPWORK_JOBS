```python
import pandas as pd
# pd.set_option('display.max_rows', 660)
```

```python
raw=pd.read_excel("./Data to be cleaned from Survey Monkey.xlsx",header=0);
raw.iloc[:,9:11]
raw.loc[:,['الحالة الأجتماعية','Unnamed: 32','Unnamed: 33','Unnamed: 34','Unnamed: 35']]
```

```python
pd.set_option('display.max_rows', 660)
Y_N=raw.iloc[1:,9:11].replace('لأ','N').replace('نعم','Y')
```

```python
import numpy as np
Y_N['Would_Participate']=Y_N.iloc[:,1].replace(np.nan,'Y') 
```

```python
#M_F male female
# M_F['Male','Female']=
# M_F['Female']=['Unnamed: 22']
# raw['الجنس']
# M_F
M_F=raw.loc[1:,['الجنس','Unnamed: 22']]
M_F['Gender']=raw.loc[1:,'الجنس'].replace('ذكر','M').replace(np.nan,'F')
M_F.rename({'الجنس':'Male','Unnamed: 22':'Female'},axis='columns',inplace=True)
M_F
```

```python
#"Kuwaiti" or "Non-Kuwaiti"
K_NK=raw.loc[1:,['الجنسية','Unnamed: 24']]
K_NK['Nationality']=raw.loc[1:,'الجنسية'].replace('كويتي','Kuwaiti').replace(np.nan,'Non-Kuwaiti').replace('انثي','Non')
K_NK.rename({'الجنسية':'Kuwaiti','Unnamed: 24':'Non-Kuwaiti'},axis='columns',inplace=True)
K_NK
```
